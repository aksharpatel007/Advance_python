Filtering Data

```python
df=pd.read_csv("ipl-matches.csv")
df
```

```python
RR=df[(df['Venue']=='Narendra Modi Stadium, Ahmedabad') & ( df['TossWinner']=="Rajasthan Royals")]
RR
```

```python
MI=df[(df['Venue']=='Wankhede Stadium') & ( df['WinningTeam']=="Mumbai Indians") & ( df['Player_of_Match']=="RG Sharma")]
MI
```

```python
Win1=df[(df["Margin"]>=50)]
Win1
```

```python
sow=df[(df["SuperOver"]=="Y")]
sow
```

```python
cwin=df[(df["WinningTeam"]=="Chennai Super Kings") & (df["City"]=="Kolkata")]
cwin.shape[0]
```

```python
msp=df[(df["Player_of_Match"]=="MS Dhoni") & ((df["Team1"]=="Mumbai Indians")| (df['Team2']=="Mumbai Indians"))]
msp
```

```python
# Gujarat Titans won the Toss and Elected to Bat and won the match.
GT=df[(df["TossWinner"]=="Gujarat Titans") & (df["TossDecision"]=="bat") & (df["WinningTeam"]=="Gujarat Titans")]
GT
```
