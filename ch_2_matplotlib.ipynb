```python
import seaborn as sns
data={'maths':[80,70,60],'Science':[75,65,55],'English':[90,85,70]}
df=pd.DataFrame(data,index=['Student A','Student B','Student C'])
sns.heatmap(df,vmin=50,vmax=100,cmap='YlGnBu',linewidths=2,linecolor='white',annot=True,cbar=False)     # YlGnBu viridis rocket
plt.title('Heatmap : Student Mark(index(DF) vmin vmax cmap linewidths linecolor annot cbar)')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()
G.add_node(1) # single node add
G.add_nodes_from([2,3])
G.add_nodes_from([4,7]) # multi node add with range() & [2,4]
G.add_edge(1,2) # CONNECT TWO NODES WITH BLACK LINE AND WITHOUT ARROW
G.add_edges_from([(2,3),(4,7)])

nx.draw_networkx(G,node_size=900,node_color='red',edge_color='white',width=4)
plt.show()
import networkx as nx
plt.show()
```

```python
# PB 137

import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
G.add_node(1) # single node add
G.add_nodes_from([2,3])
G.add_nodes_from([4,5]) # multi node add with range() & [2,4]
G.add_edge(1,2) # CONNECT TWO NODES WITH BLACK LINE AND WITHOUT ARROW
G.add_edges_from([(2,0),(2,5),(1,5),(2,4),(0,1),(0,3),(1,1),(1,3),(0,2),(2,1)])

nx.draw_networkx(G,node_size=900,node_color='red',edge_color='white',width=3)
plt.show()
import networkx as nx
plt.show()
```

```python
# PB-139

import networkx as nx
import matplotlib.pyplot as plt
G=nx.DiGraph()
insta={'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 'person3': [1,1,0,1,1],'person4': [1,1,1,0,0], 'person5': [1,1,0,0,0]}
insta_user=list(insta.keys())
G.add_nodes_from(insta_user)
insta_list=list(insta.values())
print(insta_list)
for i in range(len(insta_list)):
    for j in range(len(insta_list)):
        if insta_list[i][j]==1:
            G.add_edge(insta_user[i],insta_user[j])



nx.draw_networkx(G,node_size=2200,node_color='skyblue',edge_color='white',width=3)
plt.show()
```
